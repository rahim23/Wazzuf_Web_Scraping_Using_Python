```python
from bs4 import BeautifulSoup as bs 
```

```python
from urllib.request import urlopen
```

```python
url="https://wuzzuf.net/search/jobs/?q=data+analyste&a=hpb"
```

```python
client=urlopen(url)
```

```python
html=client.read()
```

```python
html
```

```python
client.close()
```

```python
soup=bs(html,"html.parser")
```

```python
soup
```

```python
countainer=soup.find_all("div",{"class":"css-pkv5jc"})
```

```python
f=open("wazuuf.csv","w") 
header="job_title,company_type,job_type\n"
f.write(header)
```

```python
for count in countainer : 
    jtitle=count.find_all("a",{"class":"css-o171kl"})
    job_title=jtitle[0].text.strip()
    
    cname=count.find_all("a",{"class":"css-17s97q8"})
    company_name=cname[0].text.strip()
    
    jtype=count.find_all("span",{"class":"css-1ve4b75"})
    job_type=jtype[0].text.strip()
    
    f.write(job_title + " , " + company_name + " , "+ job_type + "\n")
f.close()    
```
